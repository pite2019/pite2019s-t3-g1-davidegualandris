```python
import sklearn
import sklearn.datasets
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
#read and analyze boston dataset
from sklearn.datasets import load_boston
boston = load_boston()

print(boston.keys()) #it's a dictionary, let's see the keys
print('\n')
print(boston.data.shape) #506 istances with 13 columns
print('\n')
print(boston.feature_names) #name of the features
print('\n')
print(boston.DESCR) #description of the dataset
```

```python
#read into panda and print head of dataset
bos = pd.DataFrame(boston.data) #read
print(bos.head()) #print the head
print('\n')

#WE NOTICE THAT IT IS NOT SO READABLE
bos.columns = boston.feature_names #make it more readable
print(bos.head())
print('\n')

#WE NOTICE THAT THERE IS NO "PRICE" COLUMN
#add it do the dataframe
bos['PRICE'] = boston.target
print(bos.head())
print('\n')

print(bos.describe()) #statistical data of the dataset
```

```python
#histograms
#i randomly choose 2 columns, age and tax

#plotting age with different numbers of bins
hist = bos.hist(column = ['AGE'], bins=3) 
hist = bos.hist(column = ['AGE'], bins=6) 
hist = bos.hist(column = ['AGE'], bins=10) 
hist = bos.hist(column = ['AGE'], bins=25) 
hist = bos.hist(column = ['AGE'], bins=100) 

#plotting tax with different numbers of bins
hist = bos.hist(column = ['TAX'], bins=3) 
hist = bos.hist(column = ['TAX'], bins=6) 
hist = bos.hist(column = ['TAX'], bins=10) 
hist = bos.hist(column = ['TAX'], bins=25) 
hist = bos.hist(column = ['TAX'], bins=100) 
```

```python
#linear regression
#fitting the whole dataset
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

#split in features and label (PRICE is the label)
features_bos = bos.drop('PRICE', axis = 1)
class_bos = bos['PRICE']

lm = LinearRegression()

lm.fit(features_bos, class_bos)

pred_price = lm.predict(features_bos)

#predicted vs actual
plt.scatter(class_bos, pred_price)
plt.xlabel("Prices")
plt.ylabel("Predicted prices")
plt.title("Prices vs Predicted prices")

#score
print("Score: " + str(r2_score(class_bos, pred_price)))
```

```python
#fit column by column

for i, col in enumerate(features_bos):  

    dt=features_bos.iloc[:, [i]]
    
    lm.fit(dt, class_bos)

    pred_price = lm.predict(dt)
  
    #plot

    plt.plot(dt, pred_price, label=col)
    plt.ylabel("Predicted prices")
    plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=3,
           ncol=2, mode="expand", borderaxespad=0.)
  

    #score
    print("Score (" + col +"): " + str(r2_score(class_bos, pred_price)))
```

```python
for i, col in enumerate(features_bos):
    plt.figure(i)
    dt=features_bos.iloc[:, [i]]
    plt.scatter(dt, class_bos)
    plt.ylabel("Actual Values")
    plt.xlabel(col)
```

```python
#correlation
correlation_matrix = bos.corr().round(2)
sns.heatmap(data=correlation_matrix, annot=True)
```
